### 한 페이지 수집 함수

```python
def getAmazon(page):
    import requests
    from bs4 import BeautifulSoup as bts
    import pandas as pd
    import time
    
    ua = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 11_2_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'
    url = 'https://www.amazon.com/s'
    query = {'k': 'earbuds', 'page': page, 'ref': 'sr_pg_1'}
    headers = {'user-agent': ua, 'referer': 'https://www.amazon.com/'}
    
    res = requests.post(url = url, params = query, headers = headers)
    soup = bts(res.text, 'lxml')
    items = soup.select('div.s-main-slot.s-result-list.s-search-results.sg-row > div.s-result-item.s-asin')
    
    prod = getText('span.a-size-medium')
    star = getText('i.a-icon-star-small')
    review = getText('span.a-size-base')
    price = getText('span.a-offscreen')
    
    df = pd.DataFrame({'prod': prod, 'star': star, 'review': review, 'price': price})
    return df
```

### 반복문으로 10페이지 수집

```python
import os
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
from tqdm import tqdm
import time
```

```python
pages = 10
result = pd.DataFrame()
for i in tqdm(range(pages)):
    page = i + 1
    df = getAmazon(page)
    result = pd.concat([result, df], ignore_index = True)
    time.sleep(1)
```
